```python
import firebase_admin
from firebase_admin import credentials, firestore
import time
import papermill as pm
```

```python
# Initialize Firestore
cred = credentials.Certificate("cm_firebase.json")
firebase_admin.initialize_app(cred)

# Initialize Firestore client
db = firestore.client()
```

```python
def on_snapshot(col_snapshot, changes, read_time):
    """Callback function triggered when a new order is added."""
    for change in changes:
        if change.type.name == "ADDED":
            new_order = change.document.to_dict()
            order_id = change.document.id
            print(f"\n New order detected! Order ID: {order_id}")
            print(f"Order Details: {new_order}")

            # ðŸ”¹ Call AI model to generate recommendations
            run_ai_recommendation(order_id)
```

```python
def run_ai_recommendation(order_id):
    """This function will execute the AI model notebook when a new order is added."""

    input_notebook = "cm_eva_recommmendation.ipynb"
    output_notebook = f"output_{order_id}.ipynb"

    print(f"\n Running AI recommendation model for Order ID: {order_id}...")

    try:
        pm.execute_notebook(input_notebook, output_notebook, parameters={"order_id": order_id})
        print(f"Recommendation completed for Order ID: {order_id}\n")
    except Exception as e:
        print(f"Error running AI model: {str(e)}")
```

```python
collections= ["users", "orders"]  # Replace with your actual collection names

listeners = []
for col in collections:
    col_ref = db.collection(col)
    listener = col_ref.on_snapshot(on_snapshot)
    listeners.append(listener)
    
print(" Firestore Listener is running... Waiting for new orders...")
```
